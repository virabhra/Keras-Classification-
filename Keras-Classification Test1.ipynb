```python
#prepare for model
```

```python
X=df.drop('benign_0__mal_1', axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=104)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#model buliding
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#binary classification problem
```

```python
model=Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, verbose=1, epochs=700, validation_data=(X_test, y_test))
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,6))
loss.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
elstop=EarlyStopping(monitor='val_loss',mode='auto',verbose=1, patience=25, restore_best_weights=True)
```

```python
model=Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, verbose=1, epochs=700, validation_data=(X_test, y_test), callbacks=[elstop])
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,6))
loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, verbose=1, epochs=700, validation_data=(X_test, y_test), callbacks=[elstop])
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
prd=model.predict_classes(X_test)
```

```python
prd
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, prd))
```

```python
print(confusion_matrix(y_test, prd))
```
